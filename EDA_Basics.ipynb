```python
import pandas as pd

df = pd.read_csv("House Price India.csv")
```

Q1. What is the average, median, and standard deviation of house prices?

```python
print(df["Price"].mean())
print(df["Price"].median())
print(df["Price"].std())
```

Q2. Which number of bedrooms is most common?

```python
print(df["number of bedrooms"].mode()[0])
```

Q3. Check if the price data is skewed.

```python
print(df["Price"].skew())
```

Q4. What is the average price for each number of bedrooms?

```python
print(df.groupby("number of bedrooms")["Price"].mean())
```

Q5.  What is the relationship between living area and price? (Use : Area without Basement).

```python
print(df["Area of the house(excluding basement)"].corr(df["Price"]))
```

Q6. Identify any anomalies where houses have high prices but low areas.

```python
area_q25 = df["Area of the house(excluding basement)"].quantile(0.25)
price_q90 = df["Price"].quantile(0.90)

anomalies = df[
    (df["Price"] >= price_q90) &
    (df["Area of the house(excluding basement)"] <= area_q25)
]

print(anomalies)
```

Q7. Compare average price based on number of floors and houses with or without waterfront.

```python
print(df.groupby(
    ["number of floors", "waterfront present"]
)["Price"].mean())
```

Q8. Identify the minimum and maximum house price. What does this indicate?

Q9. Which location (zipcode/area) has the highest average price?

```python
print(df["Price"].min())
print(df["Price"].max())
```

Q10. Write at least 5 insights from your analysis.

```python
# Insight 1: Average price by bedrooms
bedroom_price = df.groupby('number of bedrooms')['Price'].mean()
print("\nAverage Price by Bedrooms:")
print(bedroom_price)

# Insight 2: Correlation between area and price
corr_area_price = df['Area of the house(excluding basement)'].corr(df['Price'])
print("\nCorrelation between Area and Price:", corr_area_price)

# Insight 3: Average price by waterfront presence
waterfront_price = df.groupby('waterfront present')['Price'].mean()
print("\nAverage Price by Waterfront:")
print(waterfront_price)

# Insight 4: Most expensive postal code
top_postal = df.groupby('Postal Code')['Price'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 Postal Codes by Average Price:")
print(top_postal)

# Insight 5: Price statistics
print("\nPrice Statistics:")
print(df['Price'].describe())

# Insight 6: Most common bedroom count
print("\nMost Common Bedroom Count:")
print(df['number of bedrooms'].mode()[0])

# Insight 7: Average price by floors
floor_price = df.groupby('number of floors')['Price'].mean()
print("\nAverage Price by Floors:")
print(floor_price)
```
